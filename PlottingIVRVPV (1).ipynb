```python

################# plot_sweep notes #################

# plot_sweep plots over range of runs given in rns
# plot_type="iv", "iv+rv", "pv", ...,  any combo of iv, rv, pv can be used delimited by +
# stitch_type="none" and stitch_type="JumpBuster" #the only two kinds of stitch
# can exclude/ include devices from being plotted
# plot_sweep(ibis, datadir, rns, exclude=[], include=NAMES, stitch_type="", plot_type="")

###################################################
```

```python
plot_sweep(ibis1, datadir, rns1, exclude=[], include=NAMES, stitch_type="none", plot_type="iv+rv+pv") # before fixing flux jumps
plot_sweep(ibis1, datadir, rns1, exclude=[], include=NAMES, stitch_type="JumpBuster", plot_type="iv+rv+pv") # after fixing flux jumps
```

```python
plot_sweep(ibis2, datadir, rns2, exclude=['TAMU A', 'SiC NFC2 A', 'NW A', 'NW B'], include=NAMES, stitch_type="none", plot_type="iv+rv+pv")
plot_sweep(ibis2, datadir, rns2, exclude=['TAMU A', 'SiC NFC2 A', 'NW A', 'NW B'], include=NAMES, stitch_type="JumpBuster", plot_type="iv+rv+pv")
```

```python
plot_sweep(ibis3, datadir, rns3, exclude=['TAMU B', 'SiC squares B'], include=NAMES, stitch_type="none", plot_type="iv+rv+pv")
plot_sweep(ibis3, datadir, rns3, exclude=['TAMU B', 'SiC squares B'], include=NAMES, stitch_type="JumpBuster", plot_type="iv+rv+pv")
```

```python
plot_sweep(ibis5, datadir, rns5, exclude=[], include=NAMES, stitch_type="none", plot_type="iv+rv+pv")
plot_sweep(ibis5, datadir, rns5, exclude=[], include=NAMES, stitch_type="JumpBuster", plot_type="iv+rv+pv")
```
